# Data analysis with ydata-profiling (Binder)

This notebook loads `ml01/data/DS1.csv` and shows a profiling report.  

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

# Path to the CSV relative to the repo root when opened at ml01/
# In Binder, you'll usually start at the repo root, so we use this relative path:
csv_path = Path('..') / 'data' / 'DS1.csv'
assert csv_path.exists(), f"Could not find {csv_path}. Adjust the path if you moved the file."

# Load data
df = pd.read_csv(csv_path)
print(df.shape)
df.head()
```

```python
# Extract TC values
tc_values = df["TC"].dropna()

# Plot histogram
plt.figure(figsize=(8, 6))
plt.hist(tc_values, bins=20, color="skyblue", edgecolor="black")
plt.title("Histogram of TC Values")
plt.xlabel("TC")
plt.ylabel("Frequency")
plt.grid(axis="y", alpha=0.75)
plt.show()
```
